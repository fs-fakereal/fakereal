```python
import pandas as pd
import json
```

```python
def data_load(path_to_json: str) -> pd.DataFrame:
    data: dict = {}
    df: pd.DataFrame = None

    with open(path_to_json, "r") as file:
        data = json.load(file)

    if data:
        df = pd.DataFrame(data)

    return df
```

```python
df = data_load("data/uniface_encoded.json")
```

```python
df.head()['uniface_ff']['uniface_Fake']['train']
```

```python
dat_fake = list(df['uniface_ff']['uniface_Fake']['train'].values())
dat_real = list(df['uniface_ff']['uniface_Real']['train'].values())
```

```python
combined = []
for col in range(len(dat_fake)):
    for path in dat_fake[col]['frames']:
        lst = [path
               .replace('deepfakes_detection_datasets/', 'data/'),
               1]
        combined.append(lst)
for col in range(len(dat_real)):
    for path in dat_real[col]['frames']:
        lst = [path
               .replace('deepfakes_detection_datasets/', 'data/'),
               0]
        combined.append(lst)
```

```python
df_dat = pd.DataFrame(combined, columns=['filepath', 'label'])
```
